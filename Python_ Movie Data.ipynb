```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv('/Users/ivanazhao/Desktop/Alex Project/Python Movie/movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# learn data structure

df.shape
```

```python
# Let's see if there is any missing data

for col in df.columns:                    # df.columns -- return column names
    missing = np.mean(df[col].isnull())   # boolean calculation -- true(1), Flase(0)
    print('{} - {}%'.format(col, missing))

#print(df['name'].isnull())
```

```python
# print(df[df['budget'].isnull()]) -- show rows of data with missing values in 'budget'

# Drop rows with missing values

df = df.dropna()
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64') # astype() chang data type

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['gross'].astype('int64')

df['runtime'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['year_correct'] = df['released'].astype(str).str[-20:-15]

df['year_correct'][:5]
```

```python
df.head()
```

```python
# Sort by gross in descending order

sort = df.sort_values(by=['gross'], inplace = False, ascending = False)  #inplace = True, changes the actual list itself while sorting.
              #inplace = False, will return a new sorted list without changing the original.
    
sort[:5]
```

```python
# Set to show number of rows you want to see

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False) # show distinct values of company names

df = df.drop_duplicates() 
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()

```

```python
df.head()
```

```python

# Plot budget vs gross earning using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'green'})
```

```python
# Let's start looking at correlation
```

```python
df.corr(method = 'pearson')  # only work on numerical values
           # pearson - assume linear relation, kendall, spearman - assume monotic relation, can be linear or nonlinear
```

```python
# High correlation between budget and gross - proved
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True) # annot - add number or not

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
